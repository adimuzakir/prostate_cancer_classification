Import Library

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```


Import Data

```python
data= pd.read_csv("Prostate_Cancer.csv")
data
```


Preprocessing Data

```python
data.info()
```

```python
data.isnull().sum()
```

```python
#cek data duplikat
data.duplicated().sum()
```

```python
#melakukan normalisasi data menggunakan min-max normalisasi
data['radius'] = (data['radius']-data['radius'].min())/(data['radius'].max()-data['radius'].min())
data['texture'] = (data['texture']-data['texture'].min())/(data['texture'].max()-data['texture'].min())
data['perimeter'] = (data['perimeter']-data['perimeter'].min())/(data['perimeter'].max()-data['perimeter'].min())
data['area'] = (data['area']-data['area'].min())/(data['area'].max()-data['area'].min())
```

```python
data
```

```python
#menentukan variabel independen
x = data.drop(["diagnosis_result"],axis=1)  
x.head()
```

```python
#menentukan variabel dependen
y=data["diagnosis_result"]
y.head()
```

```python
#percentage Split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.20) 
```

```python
#mengaktifkan package dan syntax untuk mengubah skala data
import sklearn
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
#import packages KNeigbhors Clasifier dari SKLearn
from sklearn.neighbors import KNeighborsClassifier
```

```python
k=8
```

```python
#mengaktifkan fungsi clasifikasi untuk knn
knn=KNeighborsClassifier (n_neighbors=k)
```

```python
#memasukkan data training
knn.fit(x_train,y_train)
```

```python
#menentukan prediksi
y_pred=knn.predict(x_test)
y_pred
```

```python
#menampilkan matriks prediksi
print(confusion_matrix(y_test,y_pred))
```

```python
#ketepatan hasil prediksi
print(classification_report(y_test,y_pred))
```

```python
#menghitung akurasi
akurasi=accuracy_score(y_test,y_pred)
akurasi
```
